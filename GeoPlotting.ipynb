Perfect [tutorial](http://gnperdue.github.io/yak-shaving/osx/python/matplotlib/2014/05/01/basemap-toolkit.html) for Basemap installation on Mac OS

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
rcParams.update({'font.size': 30})

%matplotlib inline
```

```python
m = Basemap(projection='mill', llcrnrlat=-90, urcrnrlat=90,\
           llcrnrlon=-180, urcrnrlon=180, resolution ="c") # Millar Cylindrical Projection
```

```python
plt.figure(figsize=(12,6))
m.drawcoastlines()
m.fillcontinents()
m.drawmapboundary()
plt.title("Quick Map Example")
plt.show()
```

## Adding colors

```python
# here we cut Antartics by setting left lower corner lattitude to -60
m = Basemap(projection='mill', llcrnrlat=-60, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution ="c") 

plt.figure(figsize=(12,6))

m.fillcontinents(color="#072b57", lake_color="#FFFFFF")
m.drawmapboundary(fill_color="#FFFFFF")
plt.title("Quick Map Example")
plt.show()
```

## Countries

```python
m = Basemap(projection='mill', llcrnrlat=-60, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution ="c") 

plt.figure(figsize=(12,6))

m.drawcountries()
m.drawstates()
m.fillcontinents(color="#04BAE3", lake_color="#FFFFFF")
m.drawmapboundary(fill_color="#FFFFFF")
plt.title("Countries and states")
plt.show()
```

## Rivers

```python
m = Basemap(projection='mill', llcrnrlat=-60, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution ="c") 

plt.figure(figsize=(12,6))
m.drawrivers()
m.fillcontinents(color="#04BAE3", lake_color="#FFFFFF")
m.drawmapboundary(fill_color="#FFFFFF")
plt.title("Rivers")
plt.show()
```

## Blue marble from NASA

```python
m = Basemap("mill") 

plt.figure(figsize=(20,15))
m.drawcoastlines()
m.drawmapboundary(fill_color="#FFFFFF")
m.bluemarble()
plt.title("Badass Earth in Python")
plt.show()
```

## Plotting on map

```python
m = Basemap(projection='mill', llcrnrlat=20, urcrnrlat=50, llcrnrlon=-130, urcrnrlon=-60, resolution ="c") 

plt.figure(figsize=(12,6))

m.drawcountries()
m.drawstates()
m.fillcontinents(color="#04BAE3", lake_color="#FFFFFF")
m.drawmapboundary(fill_color="#FFFFFF")


lat, lon = 29.76, - 95.36
x,y = m(lon, lat)
m.plot(x,y, "ro", markersize=20, alpha = 0.7)

lat, lon = 28.76, - 95.36
x,y = m(lon, lat)
m.plot(x,y, "go", markersize=20, alpha = 0.7)

lat, lon = 40.12, -104.4
x,y = m(lon, lat)
m.plot(x,y, "bo", markersize=10)


plt.title("Geo plotting")
plt.show()
```

```python
m = Basemap(projection='mill', llcrnrlat=20, urcrnrlat=50, llcrnrlon=-130, urcrnrlon=-60, resolution ="c") 

plt.figure(figsize=(12,6))

m.drawcountries()
m.drawstates()
m.fillcontinents(color="#04BAE3", lake_color="#FFFFFF")
m.drawmapboundary(fill_color="#FFFFFF")

lat = 30, 31, 32, 33, 34
lon = -103, -110, 107, -111, -115

lat2 = 40, 34, 42, 38, 44
lon2 = -113, -100, 102, -111, -112

x,y = m(lon, lat)
m.plot(x,y, "go", markersize=20, alpha = 0.7)

x,y = m(lon2, lat2)
m.plot(x,y, "ro", markersize=20, alpha = 0.7)

plt.title("Geo plotting")
plt.show()
```
